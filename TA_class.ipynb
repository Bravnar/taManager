```python
# TEACHING ASSISTANT CLASS

class TeachingAssistant:
    def __init__(self, name, email_address, userID):
        self.__name = name
        self.__email_address = email_address
        self.__userID = userID
        self.__isManager = False
        self.__isLMS = False
        self.__courses = []
    def get_last_name(self):
        return self.__name.split()[0].capitalize()
    def get_first_name(self):
        return self.__name.split()[1].capitalize()
    def get_fullname(self):
        return self.__name
    def get_email(self):
        return self.__email_address
    def get_id(self):
        return self.__userID
    def add_course(self, course_data):
        self.__courses.append(course_data)
    def get_isManager(self):
        return self.__isManager
    def get_isLMS(self):
        return self.__isLMS
    def set_isManager(self, value):
        self.__isManager = value
    def set_isLMS(self, value):
        self.__isLMS = value
    def get_courses(self):
        return self.__courses
    def print_details(self):
        print("-------------------------------------------------------------------")
        print(f"Name:\t\t{self.__name}")
        print(f"E-mail:\t\t{self.__email_address}")
        print(f"LMS_ID:\t\t{self.__userID}")
        print(f"Is Manager?\t{self.__isManager}")
        print(f"Is LMS Support?\t{self.__isLMS}")
        print("-------------------------------------------------------------------")
    def print_courses(self):
        counter = 0
        for course in self.get_courses():
            print(course)
            counter += 1
        if not counter:
            print(f"No courses detected for {self.__name}.")
        
```

```python
# TA MANAGER CLASS, NOT TO BE CONFUSED WITH THE ACTUAL TA MANAGER

class TAManager:
    def __init__(self):
        self.__ta_list = []
    def add_ta(self, ta):
        self.__ta_list.append(ta)
    def remove_ta(self, ta_to_remove):
        self.__ta_list.remove(ta_to_remove)
    def get_tas(self):
        return self.__ta_list
    def search_ta_by_last_name(self, last_name):
        for ta in self.__ta_list:
            if ta.get_last_name().lower().startswith(last_name.lower()):
                return ta
        return None
    def print_TA_list(self):
        for ta in self.__ta_list:
            print("-------------------------------------------------------------------")
            print(f"Name:\t\t{ta.get_fullname()}")
            print(f"E-mail:\t\t{ta.get_email()}")
            print(f"LMS_ID:\t\t{ta.get_id()}")
            print(f"Is Manager?\t{ta.get_isManager()}")
            print(f"Is LMS Support?\t{ta.get_isLMS()}")
        print("-------------------------------------------------------------------")
    def print_LMS(self):
        for ta in self.__ta_list:
            if ta.get_isLMS():
                print("-------------------------------------------------------------------")
                print(f"Name:\t\t{ta.get_fullname()}")
                print(f"E-mail:\t\t{ta.get_email()}")
                print(f"LMS_ID:\t\t{ta.get_id()}")
                print(f"Is Manager?\t{ta.get_isManager()}")
                print(f"Is LMS Support?\t{ta.get_isLMS()}")
        print("-------------------------------------------------------------------")
    def print_manager(self):
        for ta in self.__ta_list:
            if ta.get_isManager():
                print("-------------------------------------------------------------------")
                print(f"Name:\t\t{ta.get_fullname()}")
                print(f"E-mail:\t\t{ta.get_email()}")
                print(f"LMS_ID:\t\t{ta.get_id()}")
                print(f"Is Manager?\t{ta.get_isManager()}")
                print(f"Is LMS Support?\t{ta.get_isLMS()}")
        print("-------------------------------------------------------------------")
```
